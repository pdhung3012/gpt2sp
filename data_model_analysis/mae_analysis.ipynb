```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
```

## Within Project

### data

```python
df = pd.read_csv("./within_proj_data.csv")

x = [i for i in range(16)]
xticks = ["Appcelerator Studio", "Aptana Studio",
         "Bamboo", "Clover",
         "Data Management", "DuraCloud",
         "JIRA Software", "Mesos",
         "Moodle", "Mule",
         "Mule Studio", "Spring XD",
         "Talend Data Quality", "Talend ESB",
         "Titanium", "Usergrid"]

sp_maen = df["SP_mean"].tolist()
sp_std = df["SP_std"].tolist()

gpt2sp = df["GPT2SP"].tolist()
deepse = df["DeepSE"].tolist()

lstm_rf = df["LSTM+RF"].tolist()
lstm_svm = df["LSTM+SVM"].tolist()
doc2vec_rf = df["Doc2Vec+RF"].tolist()
bow_rf = df["BOW+RF"].tolist()
lstm_atlm = df["LSTM+ATLM"].tolist()
lstm_lr = df["LSTM+LR"].tolist()
```

### Plot - Simple

```python
plt.plot(x, gpt2sp, label='MAE of GPT2SP')
plt.plot(x, deepse, label='MAE of DeepSE')
plt.plot(x, sp_mean, label='SP Mean')
plt.plot(x, sp_std, label='SP STD')
plt.xticks(np.arange(min(x), max(x)+1, 1.0))

ax = plt.gca()
ax.legend(fontsize='15')
ax.set_xticklabels(xticks, rotation=90, ha='left')

fig = plt.gcf()
fig.set_size_inches(10, 10)
fig.savefig('test2png.png', dpi=100)
```
